#### 级联操作
    - 将多个numpy数组进行横向或者纵向的拼接
- axis轴向的理解
    - 0：列
    - 1：行

```python
arr
```

```python
np.concatenate((arr, arr), axis=1)
```

```python
arr_2 = np.random.randint(0, 100, size=(3, 7))
arr_3 = np.random.randint(0, 100, size=(3, 6))
np.concatenate((arr_2, arr_3), axis=1)  # 横向拼接
```

```python
plt.imshow(np.concatenate((img_arr, img_arr, img_arr), axis=1))  # 纵向拼接
```
