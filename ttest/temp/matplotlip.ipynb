```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1, 2, 3, 4, 5])
y = x + 3

plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.plot(x + 1, y - 2)
```

```python
plt.plot(x, y, x+1, y-2)
```

```python
plt.figure(figsize=(5, 8))
plt.plot(x, y)
```

```python
plt.plot(x, y, label='x, y')
plt.plot(x+1, y-2, label='x+1, y-2')
plt.legend(ncol=2)
```

```python
plt.plot(x, y)
plt.xlabel('temp')
plt.ylabel('dist')
plt.title('dist&temp')
```

```python
fig = plt.figure()
plt.plot(x, y, label='x, y')
fig.savefig('./123.png')
```
